```python
#Import libreries

import pandas as pd
from cleaner import categorical_to_numeric
```

# Preparing the data

Data train

```python
#Loading CSV
data = pd.read_csv("Input/diamonds_train.csv")
```

```python
#Imported function from cleaner.py file 
categorical_to_numeric(data)
#Rename
data.rename(columns={'Unnamed: 0': 'id'}, inplace=True)
#Removing somo columns
data.drop(['x', 'y', 'z'], axis=1, inplace=True)
data.head()
```

Data test

```python
#Test dataset
data_test = pd.read_csv('./input/diamonds_test.csv')

#Imported function from cleaner.py file 
categorical_to_numeric(data_test)
#Rename
data_test.rename(columns={'Unnamed: 0': 'id'}, inplace=True)
#Removing somo columns
data_test.drop(['x', 'y', 'z'], axis=1, inplace=True)
data_test.head()
```

```python
#H2O
import h2o
from h2o.automl import H2OAutoML
```

```python
#Initiating
h2o.init()
```

```python
#Defining x and y
x =['id', 'carat', 'cut', 'color', 'clarity', 'depth', 'table']
y = "price"
```

```python
h2_train = h2o.H2OFrame(data)
h2_test = h2o.H2OFrame(data_test)
```

# Getting best models

```python
#Getting the best models
automl = H2OAutoML(max_models=30, seed=42,max_runtime_secs=3600, sort_metric='RMSE')
automl.train(x=x, y=y, training_frame=h2_train)
```

```python
#best models
leader_board = automl.leaderboard
leader_board.head()
```
